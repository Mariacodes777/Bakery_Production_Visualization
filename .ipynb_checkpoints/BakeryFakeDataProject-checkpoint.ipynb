 I am really excited to make this fabricated dataset just for fun to show how something like this could be used to analyize sales patterns. It would help with creating more accurate production logs with the hopes of increasing sales, reducing sold out periods so items are more readily available, and overall, creating a more streamlined production process!

### Enjoy!

```python
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Defining the bakery items and the days of the week. Going to weigh more heavily Friday, Saturday, and 
# Sunday since we know those are the busiest days.
```

```python
Bakery_items = ['FruitCakeSm', 'FruitCakeLg', 'Blk&Wht', 'ChocRasp', 'VanillaLG', 'VanillaSm', 'DreamSm', 'DreamLg', 'ChzCake', 'FruitTarts', 'ChocMousse', 'SheetVan', 'SheetChoc', 'Carrot', 'VeganOreo', 'VeganVan']
```

```python
Days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
# Creating an empty list to store the data
```

```python
Data = []
```

```python
# Creating a 1000 rows of random data
for i in range(1000):
    Bakery_item = random.choice(Bakery_items)
    # Weighing the selection of days to favor busy days like Friday, Saturday, and Sunday.
    # Below you see I am creating this data with Monday to Thursday being the same in busyness, weighing Friday
    # as busier, Saturday as the busiest, and Sunday as busy but less than Saturday.
    Day = random.choices(Days, weights=[1, 1, 1, 1, 2, 4, 3])[0]
    Data.append((Bakery_item, Day))
```

```python
# Converting the list to a pandas DataFrame
BakeryData = pd.DataFrame(Data, columns=['BakeryItem', 'DayofWeek'])
```

```python
OrderOfDays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize = (17, 8))
sns.histplot(data = BakeryData, x = 'BakeryItem', hue='DayofWeek', hue_order=OrderOfDays, multiple= 'dodge')

bake_items = BakeryData['BakeryItem'].unique()
for i in range(1, len(bake_items)):
    plt.axvline(x=i - 0.5, color='black', linestyle='--')

plt.show()
```

```python
# Above is another fancy way to look at each item sold by day of the week to be able to compare and contrast.
```

```python
# To get a look at each day of the week individually:
```

```python
# Group the data by day of the week and bakery item
SoldItems = BakeryData.groupby(['DayofWeek', 'BakeryItem']).size().reset_index(name='counts')

# View Each day of the week individually
days = SoldItems['DayofWeek'].unique()

# Seting the order of the days of the week
DaysInOrder = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for day in DaysInOrder:
    day_SoldItems = SoldItems[SoldItems['DayofWeek'] == day]

    # Sorting the data in ascending order of the number of items sold
    day_SoldItems = day_SoldItems.sort_values(by='counts', ascending=False)
    
    fig, ax = plt.subplots()

    # Creating the plot
    day_SoldItems.plot(kind='bar', x='BakeryItem', y='counts', ax=ax)

    # Seting labels and title
    ax.set_xlabel('Bakery Item')
    ax.set_ylabel('Number of Items Sold')
    ax.set_title(f'Number of Bakery Items Sold on {day}')

    # Visualizing plots
    plt.show()
```

```python
# Now that you can see the items sold per day, it would be able to serve as a reference if this was going to be 
# used in a real bakery to help create production logs based off of sales per day and shelf life. For example,
# for cakes with a 5 day shelf life, the average number sold between the 5 days could be reflected through 
# production. I hope you find this a fun and useful way to visualize data!
```
